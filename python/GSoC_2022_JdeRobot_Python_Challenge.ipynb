```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

x = 5 * np.random.random_sample()
y = 5 * np.random.random_sample()

vx = np.random.random_sample() / 5
vy = np.random.random_sample() / 5
```

```python
def frame(f):
    ax.clear()
    global x, y, vx, vy
    if x < 0.1 or x >= 4.9:
        vx *= -1
    if y < 0.1 or y >= 4.9:
        vy *= -1
    x += vx
    y += vy

    plt.title("Brownian Motion")
    ax.set_xlim(0,5)
    ax.set_ylim(0,5)
    plot = ax.scatter(x, y, c='b')
    return plot
```

```python
anim = animation.FuncAnimation(fig, frame, frames=200, blit=False, repeat=True, interval=50)
```
